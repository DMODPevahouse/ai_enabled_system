For creating yolo from voc I was able to kinda sort together this implementation to take in the voc file and create the needed text file for yolo. I did not get a chance to look at to many libraries for it, but this worked well enough.

```python
import os
import pylabel
import xml.etree.ElementTree as ET

def convert_voc_to_yolo(voc_annotations_dir, output_dir):
    # Iterate through all the XML files in the VOC annotations directory
    for filename in os.listdir(voc_annotations_dir):
        if filename.endswith(".xml"):
            # Parse the XML file
            tree = ET.parse(os.path.join(voc_annotations_dir, filename))
            root = tree.getroot()

            # Initialize variables for storing image width, height, and object annotations
            image_width = int(root.find("size/width").text)
            image_height = int(root.find("size/height").text)
            object_annotations = []

            # Iterate through all the object annotations in the XML file
            for obj in root.findall("object"):
                # Get the object's bounding box coordinates
                xmin = int(obj.find("bndbox/xmin").text)
                ymin = int(obj.find("bndbox/ymin").text)
                xmax = int(obj.find("bndbox/xmax").text)
                ymax = int(obj.find("bndbox/ymax").text)

                # Convert the bounding box coordinates to YOLO format
                x_center = (xmin + xmax) / (2 * image_width)
                y_center = (ymin + ymax) / (2 * image_height)
                width = (xmax - xmin) / image_width
                height = (ymax - ymin) / image_height

                # Store the object annotation in a list
                object_annotations.append([obj.find("name").text, x_center, y_center, width, height])

            # Write the object annotations to a text file in YOLO format
            with open(os.path.join(output_dir, filename.replace(".xml", ".txt")), "w") as f:
                for annotation in object_annotations:
                    f.write(f"{annotation[0]} {annotation[1]} {annotation[2]} {annotation[3]} {annotation[4]}\n")
```

```python
voc_annotations_dir = "images"
output_dir = "updated_images/yolo"

convert_voc_to_yolo(voc_annotations_dir, output_dir)
```

Given more time I would probably spend a larger amount of time creating, or finding a better library, to do this, but this worked for this assignment. The json file is printed at the bottom to save space here.

```python
import os
import xml.etree.ElementTree as ET
import json

def pascal_voc_to_json(xml_dir, output_file):
    """
    Converts all Pascal VOC XML files in a directory to a single JSON file.

    Parameters:
    xml_dir (str): The directory containing the Pascal VOC XML files.
    output_file (str): The path to the output JSON file.

    Returns:
    None
    """

    # Initialize lists for annotations and images
    annotations = []
    images = []

    # Iterate through XML files in the directory
    for filename in os.listdir(xml_dir):
        if filename.endswith(".xml"):
            file_path = os.path.join(xml_dir, filename)
            tree = ET.parse(file_path)
            root = tree.getroot()

            # Extract image information
            image_width = int(root.find("size/width").text)
            image_height = int(root.find("size/height").text)
            image_filename = root.find("filename").text

            # Extract annotation information
            for obj in root.findall("object"):
                annotation = {
                    "name": obj.find("name").text,
                    "pose": obj.find("pose").text,
                    "truncated": int(obj.find("truncated").text),
                    "difficult": int(obj.find("difficult").text),
                    "category_id": 0,
                    "bndbox": {
                        "xmin": int(obj.find("bndbox/xmin").text),
                        "ymin": int(obj.find("bndbox/ymin").text),
                        "xmax": int(obj.find("bndbox/xmax").text),
                        "ymax": int(obj.find("bndbox/ymax").text)
                    }
                }
                annotations.append(annotation)

            # Add extracted information to the respective lists
            images.append({
                "file_name": image_filename,
                "height": image_height,
                "width": image_width,
                "id": len(images)
            })

    # Create the JSON output
    output_data = {
        "images": images,
        "annotations": annotations,
        "categories": [
            {
                "id": 1,
                "name": "license_plate"
            }
        ]
    }

    with open(output_file, "w") as outfile:
        json.dump(output_data, outfile)

```

```python
xml_dir = "images"
output_file = "updated_images/coco/output.json"
pascal_voc_to_json(xml_dir, output_file)
```

```python
import subprocess

def find_tesseract_executable():

    command = "which"

    try:
        output = subprocess.check_output([command, "pytesseract"], stderr=subprocess.STDOUT, shell=True)
        return output.decode().strip()
    except (subprocess.CalledProcessError, FileNotFoundError) as e:
        return None
```

```python
import matplotlib.pyplot as plt
import cv2
import pytesseract
from PIL import Image

def ocr_license_plate(image):
    """
    Extracts the license plate number using Google's Tesseract OCR.

    Parameters:
    image (PIL.Image): The cropped image.

    Returns:
    str: The extracted license plate number.
    """

    pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'
    text = pytesseract.image_to_string(image, config="--psm 6")
    return text.strip()
```

```python
def crop_license_plate(image_path, bbox):
    """
    Crops the license-plate-only portion of the image.

    Parameters:
    image_path (str): The path to the image file.
    bbox (list): The list of bounding boxes.

    Returns:
    Image: The cropped image.
    """

    image = cv2.imread(image_path)
    xmin, ymin, xmax, ymax = bbox['xmin'], bbox['ymin'], bbox['xmax'], bbox['ymax']
    cropped_image = image[ymin:ymax, xmin:xmax]
    try:  # Check if the cropped image is not empty
        return Image.fromarray(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
    except Exception as e:
        print("Error: The cropped image is empty.")
        return None
```

```python
# Function to load bounding box coordinates from a COCO-style JSON file
def load_coco_bbox(json_file, n):
    with open(json_file, 'r') as f:
        data = json.load(f)
    bbox = data['annotations'][n]['bndbox']
    return bbox

# Example usage
image_path = "images/N1.jpeg"
json_file = "updated_images/coco/output.json"  # Path to the COCO-style JSON file

# Load bounding box coordinates from the JSON file
bbox = load_coco_bbox(json_file, 0)

# Display the original image
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.show()

# Crop the license plate
cropped_image = crop_license_plate(image_path, bbox)

# Display the cropped image
plt.figure(figsize=(10, 10))
plt.imshow(cropped_image)
plt.title("Cropped Image")
plt.show()
# Extract the license plate number
license_plate_number = ocr_license_plate(cropped_image)
print("License Plate Number: ", license_plate_number)
```

```python
def load_name(json_file, n):
    with open(json_file, 'r') as f:
        data = json.load(f)
    name = data['images'][n]['file_name']
    return name
```

Below loops through every image that the data is available on and attempts to read the license plate for every image. 

```python
image_path = "images/"
json_file = "updated_images/coco/output.json"  # Path to the COCO-style JSON file
i = 0
# Load bounding box coordinates from the JSON file
for picture in os.listdir(image_path):
    name = load_name(json_file, i)
    if picture.endswith('.jpeg') or picture.endswith('.jpg') and f'{os.path.isfile(os.path.basename([name]))}.xml':
        print(i, name)
        image_directory = os.path.join(image_path, picture)
        bbox = load_coco_bbox(json_file, i)
        # Crop the license plate
        cropped_image = crop_license_plate(image_directory, bbox)
        if cropped_image is not None:
        # Extract the license plate number
            license_plate_number = ocr_license_plate(cropped_image)
            print("License Plate Number: ", license_plate_number)
            i += 1
        else:
            print("license plate not found, moving on")
    if i > 224:
        break
```
