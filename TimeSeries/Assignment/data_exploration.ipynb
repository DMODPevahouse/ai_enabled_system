```python
# Convert trans_date to datetime if it's not already
df['trans_date'] = pd.to_datetime(df['trans_date'])

# Create a new column 'trans_week' that represents the week of the trans_date
df['trans_week'] = df['trans_date'].dt.to_period('W')

# Now, group by trans_week instead of trans_date
df['transaction_count'] = df.groupby('trans_week').transform('size')
df['fraud_count'] = df['is_fraud'].groupby(df['trans_week']).transform('sum')
df['non_fraud_count'] = df['transaction_count'] - df['fraud_count']

# Select the relevant columns
result_df = df[['trans_week', 'transaction_count', 'fraud_count', 'non_fraud_count']]

# Sort by trans_week
result_df = result_df.sort_values(by='trans_week', ascending=True)

# Set trans_week as the index
result_df = result_df.set_index("trans_week")

# Drop duplicates
result_df = result_df.drop_duplicates()
```

```python
# Create a new column that contains the year and month of the transaction date
df['transaction_month'] = df['trans_date'].dt.to_period('M')

# Group the data by transaction month and calculate the size of each group
df['transaction_count'] = df.groupby('transaction_month').transform('size')

# Calculate the number of fraudulent transactions for each transaction month
df['fraud_count'] = df['is_fraud'].groupby(df['transaction_month']).transform('sum')

# Calculate the number of non-fraudulent transactions for each transaction month
df['non_fraud_count'] = df['transaction_count'] - df['fraud_count']

# Select only the columns of interest
result_df = df[['transaction_month', 'transaction_count', 'fraud_count', 'non_fraud_count']]

# Sort the data by transaction month and reset the index
result_df = result_df.sort_values(by='transaction_month', ascending=True).reset_index(drop=True)

# Set the transaction month as the index of the result dataframe
result_df = result_df.set_index("transaction_month")

# Remove any duplicate rows
result_df = result_df.drop_duplicates()
```

```python
train = df_weekly.iloc[:265-26]
test = df_weekly.iloc[265-26:]
```

```python
train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
# FIT ONLY TRAIN DATA, OTHERWISE LEAKS SOME OF THE TEST DATA
scaler.fit(train)

scaled_train = scaler.transform(train)
scaled_test = scaler.transform(test)
```

```python
from keras.preprocessing.sequence import TimeseriesGenerator
n_input = 26
generator = TimeseriesGenerator(scaled_train, scaled_train, length=n_input, batch_size=1)
```

```python
n_features = 1
```

```python
generator[0]
```

```python
generator[0][0].shape
```

```python
generator[0][1].shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM

model = Sequential()
model.add(LSTM(100, activation='relu', input_shape=(n_input, n_features)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

model.summary()
```

```python
model.fit_generator(generator,epochs=20)
```

```python
import matplotlib.pyplot as plt

loss_per_epoch = model.history.history['loss']
plt.plot(range(len(loss_per_epoch)),loss_per_epoch)
```

```python
first_eval_batch = scaled_train[-26:]
first_eval_batch = first_eval_batch.reshape((1, n_input, n_features))
model.predict(first_eval_batch)

first_eval_batch = scaled_train[-n_input:]
current_batch = first_eval_batch.reshape((1, n_input, n_features))
```

```python
current_batch.shape
```

```python
np.append(current_batch[:,1:,:],[[[99]]],axis=1)
```

```python
test_predictions = []

first_eval_batch = scaled_train[-n_input:]
current_batch = first_eval_batch.reshape((1, n_input, n_features))
# go beyond len(test) to go into the unknown future (no way to measure success except wait!)
for i in range(len(test)):
    
    # get prediction 1 time stamp ahead ([0] is for grabbing just the number instead of [array])
    current_pred = model.predict(current_batch)[0]
    
    # store prediction
    test_predictions.append(current_pred) 
    
    # update batch to now include prediction and drop first value
    current_batch = np.append(current_batch[:,1:,:],[[current_pred]],axis=1)
```

```python
true_predictions = scaler.inverse_transform(test_predictions)
test['Predictions'] = true_predictions
```
