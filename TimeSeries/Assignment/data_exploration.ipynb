```python
# Assuming the dataframe is called 'df' and has the 'trans_date' and 'is_fraud' columns

# Convert the 'trans_date' column to datetime
df['trans_date'] = pd.to_datetime(df['trans_date'])

# Group the data by month and calculate the fraud and non-fraud totals
df_monthly = df.groupby(df['trans_date'].dt.strftime('%Y-%m')).agg({
    'is_fraud': ['sum', 'count']
}).reset_index()

# Rename the columns
df_monthly.columns = ['month', 'fraud_total', 'non_fraud_total']

# Display the result
df_monthly
```

```python
df['trans_date'] = pd.to_datetime(df['trans_date'])

# Group the data by week and calculate the fraud and non-fraud totals
df_weekly = df.groupby(df['trans_date'].dt.strftime('%Y-%W')).agg({
    'is_fraud': ['sum', 'count']
}).reset_index()

# Rename the columns
df_weekly.columns = ['week', 'fraud_total', 'non_fraud_total']

# Display the result
df_weekly
```

```python
# assuming df is your dataframe and 'Transaction Date' is the column containing the transaction dates
df['trans_date'] = pd.to_datetime(df['trans_date'])

# create a new column called 'Transaction Count' that counts each transaction in a month
df['count'] = df.groupby(df['trans_date'].dt.strftime('%Y-%W')).cumcount() + 1

# transform the data to have a month and a count of transactions
df_weekly = df.groupby(df['trans_date'].dt.strftime('%Y-%W'))['count'].count().reset_index()
```

```python
len(df_weekly)
```

```python
train = df_weekly.iloc[:265-26]
test = df_weekly.iloc[265-26:]
```

```python
train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
# FIT ONLY TRAIN DATA, OTHERWISE LEAKS SOME OF THE TEST DATA
scaler.fit(train)

scaled_train = scaler.transform(train)
scaled_test = scaler.transform(test)
```

```python
from keras.preprocessing.sequence import TimeseriesGenerator
n_input = 26
generator = TimeseriesGenerator(scaled_train, scaled_train, length=n_input, batch_size=1)
```

```python
n_features = 1
```

```python
generator[0]
```

```python
generator[0][0].shape
```

```python
generator[0][1].shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM

model = Sequential()
model.add(LSTM(100, activation='relu', input_shape=(n_input, n_features)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

model.summary()
```

```python
model.fit_generator(generator,epochs=20)
```

```python
import matplotlib.pyplot as plt

loss_per_epoch = model.history.history['loss']
plt.plot(range(len(loss_per_epoch)),loss_per_epoch)
```

```python
first_eval_batch = scaled_train[-26:]
first_eval_batch = first_eval_batch.reshape((1, n_input, n_features))
model.predict(first_eval_batch)

first_eval_batch = scaled_train[-n_input:]
current_batch = first_eval_batch.reshape((1, n_input, n_features))
```

```python
current_batch.shape
```

```python
np.append(current_batch[:,1:,:],[[[99]]],axis=1)
```

```python
test_predictions = []

first_eval_batch = scaled_train[-n_input:]
current_batch = first_eval_batch.reshape((1, n_input, n_features))
# go beyond len(test) to go into the unknown future (no way to measure success except wait!)
for i in range(len(test)):
    
    # get prediction 1 time stamp ahead ([0] is for grabbing just the number instead of [array])
    current_pred = model.predict(current_batch)[0]
    
    # store prediction
    test_predictions.append(current_pred) 
    
    # update batch to now include prediction and drop first value
    current_batch = np.append(current_batch[:,1:,:],[[current_pred]],axis=1)
```

```python
true_predictions = scaler.inverse_transform(test_predictions)
test['Predictions'] = true_predictions
```

```python
test.plot(figsize=(12,8))
```
